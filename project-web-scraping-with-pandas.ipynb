## TL; DR

在這個專案中，我們打算使用 Pandas 中的函式 read_html 將 HTML 文件中的 <table\></table\> 資料擷取成為 DataFrame，以 [Yahoo! 奇摩股市](https://tw.stock.yahoo.com/)的兩個頁面為例。

## 上市行情類排行榜：單日成交價排行的頁面

<https://tw.stock.yahoo.com/d/i/rank.php?t=pri&e=TAI&n=100>

```python
rank_url = "https://tw.stock.yahoo.com/d/i/rank.php?t=pri&e=TAI&n=100"
html_tables = pd.read_html(rank_url, encoding='big5')
print(len(html_tables)) # 頁面中有幾組 <table></table> 標記
```

```python
html_tables[4] # 所需要的資料框
```
